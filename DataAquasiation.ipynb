```python
res = requests.get('https://webscraper.io/test-sites/tables')
```

```python
res
```

```python
res.text
```

```python
soup = bs4.BeautifulSoup(res.text)
print(soup.prettify())
```

```python
table_tag = soup.find_all('table')
```

```python
table_tag
```

```python
table_data = []
header = True
for tb in table_tag:
    
    if header == True:
        row = []
        for th in tb.find_all('th'):
            row.append(th.text)
        header = False
        table_data.append(row)
        
    for tr in tb.find_all('tr'):
        row = []
        for td in tr.find_all('td'):
            row.append(td.text)
        table_data.append(row)
```
