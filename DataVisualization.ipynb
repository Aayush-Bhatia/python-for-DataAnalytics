# subplot use hota h page o diff. parts m divide krke kaam krne k liye
- isme 4 chize deni hoti h (row, col, index, )

```python
plt.figure(figsize=(12,6))
x = np.linspace(-2,2,1000)

plt.subplot(2,2,1)

y = np.sin(2*np.pi*x)

plt.plot(x,y)
plt.grid()
plt.title('sin wave')


plt.subplot(2,2,2)
#x = np.linspace(-2,2,1000)
y = np.cos(2*np.pi*x)

#plt.figure(figsize=(10,2))
plt.plot(x,y)
plt.grid()
plt.title('cos wave')


plt.subplot(2,2,3)
#x = np.linspace(-2,2,1000)
y = np.tan(2*np.pi*x)

#plt.figure(figsize=(10,2))
plt.ylim(-10,10)
plt.plot(x,y)
plt.grid()

plt.title('tan wave')


plt.subplot(2,2,4)
#x = np.linspace(-2,2,1000)
y = 1/np.tan(2*np.pi*x)

#plt.figure(figsize=(10,2))
plt.ylim(-10,10)
plt.plot(x,y)
plt.grid()

plt.title('cot wave')

plt.suptitle('subploting')
plt.subplots_adjust(hspace = 0.3)
plt.show()
```

```python
x = np.linspace(-2,2,1000)
ysin = np.sin(2*np.pi*x)
ycos = np.cos(2*np.pi*x)
ytan = np.tan(2*np.pi*x)
ycot = 1/np.tan(2*np.pi*x)


fig, axes = plt.subplots(2,2,figsize = (12,6))

axes[0,0].plot(x,ysin)
axes[0,0].grid()
#axes[0,0].set_ylim(-5,5)
axes[0,0].set_title('sin wave')

axes[0,1].plot(x,ycos)
axes[0,1].grid()
#axes[0,1].set_ylim(-5,5)
axes[0,1].set_title('cos wave')

axes[1,0].plot(x,ytan)
axes[1,0].grid()
axes[1,0].set_ylim(-5,5)
axes[1,0].set_title('tan wave')

axes[1,1].plot(x,ycot)
axes[1,1].grid()
axes[1,1].set_ylim(-5,5)
axes[1,1].set_title('cot wave')

plt.suptitle('trignometry')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Let's choose a feature (e.g., 'sepal_length')
data = iris['sepal_length']

# Step 1: Sort the data
sorted_data = np.sort(data)

# Step 2: Calculate the CDF values
cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)

# Step 3: Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(sorted_data, cdf, marker='.', linestyle='none')
plt.title('CDF of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Let's choose a feature (e.g., 'sepal_length')
data = iris['petal_length']

# Step 1: Sort the data
sorted_data = np.sort(data)

# Step 2: Calculate the CDF values
cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)

# Step 3: Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(sorted_data, cdf, marker='.', linestyle='none')
plt.title('CDF of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```
